```python
import gensim
import pandas as pd
```

```python
data = pd.read_json('Cell_Phones_and_Accessories_5.json', lines=True)
data
```

```python
data_review = data.reviewText.apply(gensim.utils.simple_preprocess)
data_review
```

```python
model = gensim.models.Word2Vec(
window=2,
min_count=2,
workers=5)
```

```python
model.build_vocab(data_review, progress_per=1000)
```

```python
model.train(data_review, total_examples=model.corpus_count, epochs=50)
```

```python
model.wv.similarity(w1='great', w2='legit')
```

```python
df = pd.read_json('Sports_and_Outdoors_5.json', lines=True)
df
```

```python
text = df.reviewText.apply(gensim.utils.simple_preprocess)
text
```
